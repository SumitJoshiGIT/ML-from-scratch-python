```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

So assuming a n-dimensional array I 

```python
from sklearn.datasets import load_iris

X=np.array([[1,2,3],[4,5,2],[9,1,2],[7,5,6],[4,0,2]])
Y=np.zeros(X.shape[0])
# Load the iris dataset
iris = load_iris()
X = iris.data
K=2 #X is the data, K is the number of clusters
N=10#iterations
delta_Y=0.1
print(X)
```

```python
centroids = X[np.random.choice(X.shape[0], K, replace=False)]    

while(N>0):
    sub = np.array([abs(x-centroids) for x in X])
    manhatten = np.sum(sub, axis=2)
    Y = np.argmin(manhatten, axis=1)
    plt.figure()
    for k in range(K):
        plt.scatter(X[Y == k][:, 0], X[Y == k][:, 1])
        mean = np.mean(X[Y == k], axis=0)
        diff = abs(centroids[k] - mean).mean()
        if(diff > delta_Y):
            centroids[k] = mean 
        plt.scatter(centroids[k][0], centroids[k][1], marker='x', s=200, c='red')
    N -= 1
    plt.show()

print(centroids)
```
